```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Load the data from the Excel file
df = pd.read_excel(r'D:\Merge file.xlsx')

# Extract and convert the 'Comment' column to strings
text_column = df['Comment'].astype(str)

# Tokenize and preprocess text
def preprocess_text(text):
    words = word_tokenize(text)
    return words

word_lengths = [[] for _ in range(8)]  # Create lists to store words of lengths 0 to 7

for comment in text_column:
    words = preprocess_text(comment)
    for word in words:
        length = len(word)
        if 1 <= length <= 7:  # Only consider words with lengths from 1 to 7
            word_lengths[length].append(word)

top_10_words = [[] for _ in range(8)]  # Create lists to store the top 10 frequent words of lengths 0 to 7

for i in range(1, 8):
    word_freq = nltk.FreqDist(word_lengths[i])
    top_10_words[i] = word_freq.most_common(10)

# Create a Pandas DataFrame to store the results
results = {'Word Length': [], 'Top Words': []}
for i in range(1, 8):
    results['Word Length'].append(f'{i}-letter')
    top_words_str = ', '.join([f'{word} ({freq})' for word, freq in top_10_words[i]])
    results['Top Words'].append(top_words_str)

df_results = pd.DataFrame(results)

# Save the results as an Excel file
output_excel_path = 'Top_Words_By_Length.xlsx'
df_results.to_excel(output_excel_path, index=False)
print(f"Top words by length saved to {output_excel_path}")
```

```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
nltk.download('punkt')
import random  # Import the random module

# Set a random seed for reproducibility
random.seed(42)

df = pd.read_excel('D:\Merge file.xlsx')

text_column = df['Comment']
text_column = text_column.apply(lambda x: str(x))

# Tokenize and preprocess text
def preprocess_text(text):
    words = word_tokenize(text)
    return words

word_lengths = [[] for _ in range(8)]  # Create lists to store words of lengths 0 to 7

for comment in text_column:
    words = preprocess_text(comment)
    for word in words:
        length = len(word)
        if 1 <= length <= 7:  # Only consider words with lengths from 1 to 7
            word_lengths[length].append(word)

total_words = sum(len(word_lengths[i]) for i in range(1, 8))

# Calculate the percentage of the top 10 frequent n-letter words for n = 1 to 7
top_10_percentages = [[] for _ in range(8)]
for i in range(1, 8):
    word_freq = nltk.FreqDist(word_lengths[i])
    top_10_words = word_freq.most_common(10)
    for word, freq in top_10_words:
        percentage = (freq / total_words) * 100
        top_10_percentages[i].append((word, percentage))

# Print the top 10 frequent n-letter words in percentage for n = 1 to 7
for i in range(1, 8):
    print(f"Top 10 frequent {i}-letter words in percentage:")
    for word, percentage in top_10_percentages[i]:
        print(f"{word}: {percentage:.2f}%")
    print()
```
