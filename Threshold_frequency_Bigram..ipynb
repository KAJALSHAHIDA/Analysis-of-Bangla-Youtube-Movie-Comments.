```python
import nltk
from nltk import FreqDist
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import pandas as pd
nltk.download('punkt')

df = pd.read_excel('D:\Merge file.xlsx')

text_column = df['Comment']
text_column = text_column.apply(lambda x: str(x))

def is_valid_bengali_word(word):
    bengali_range = (0x0980, 0x09FF)
    return all(ord(char) >= bengali_range[0] and ord(char) <= bengali_range[1] for char in word)

def preprocess_text(text):
    words = word_tokenize(text)
    bengali_words = [word for word in words if is_valid_bengali_word(word)]
    return bengali_words

bengali_bigrams = []

for comment in text_column:
    bengali_words = preprocess_text(comment)
    comment_bigrams = list(ngrams(bengali_words, 2))  # Generate bigrams
    bengali_bigrams.extend(comment_bigrams)

fdist = FreqDist(bengali_bigrams)

# Set threshold frequencies
thresholds = [20, 50, 100, 200]

for threshold in thresholds:
    # Filter bigrams by frequency threshold
    filtered_bigrams = {bigram: freq for bigram, freq in fdist.items() if freq >= threshold}
    
    # Create a DataFrame from the filtered bigrams
    output_data = {'Bigram': filtered_bigrams.keys(), 'Frequency': filtered_bigrams.values()}
    output_df = pd.DataFrame(output_data)
    
    # Sort the DataFrame by frequency in descending order
    output_df = output_df.sort_values(by='Frequency', ascending=False)
    
    # Save the filtered bigrams to an Excel file with the threshold in the filename
    output_excel_path = f'Bigram_output_threshold_{threshold}.xlsx'
    output_df.to_excel(output_excel_path, index=False)
    print(f"Filtered Bigram output (Threshold {threshold}) saved to", output_excel_path)
```
