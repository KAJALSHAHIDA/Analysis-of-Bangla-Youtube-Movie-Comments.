```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
nltk.download('punkt')

df = pd.read_excel('D:\Merge file.xlsx')

text_column = df['Comment']
text_column = text_column.apply(lambda x: str(x))

# Tokenize and preprocess text
def preprocess_text(text):
    words = word_tokenize(text)
    return words

initial_letter_freq = {}  # Initialize a dictionary to store initial letter frequencies

for comment in text_column:
    words = preprocess_text(comment)
    for word in words:
        if word and '\u0980' <= word[0] <= '\u09FF':  # Check if the first character is a Bengali letter
            initial_letter = word[0].lower()  # Convert to lowercase for case-insensitive counting
            if initial_letter in initial_letter_freq:
                initial_letter_freq[initial_letter] += 1
            else:
                initial_letter_freq[initial_letter] = 1

# Calculate total initial letter count
total_initial_letters = sum(initial_letter_freq.values())

# Calculate and print the percentage of occurrence of each initial letter
for initial_letter, freq in initial_letter_freq.items():
    percentage = (freq / total_initial_letters) * 100
    print(f"Initial Letter: {initial_letter}, Occurrences: {freq}, Percentage: {percentage:.2f}%")
```

```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Read the Excel file
df = pd.read_excel(r'D:\Merge file.xlsx')

# Ensure the 'Comment' column is treated as strings
df['Comment'] = df['Comment'].astype(str)

# Tokenize and preprocess text
def preprocess_text(text):
    words = word_tokenize(text)
    return words

initial_letter_freq = {}  # Initialize a dictionary to store initial letter frequencies

for comment in df['Comment']:
    words = preprocess_text(comment)
    for word in words:
        if word and '\u0980' <= word[0] <= '\u09FF':  # Check if the first character is a Bengali letter
            initial_letter = word[0].lower()  # Convert to lowercase for case-insensitive counting
            initial_letter_freq[initial_letter] = initial_letter_freq.get(initial_letter, 0) + 1

# Calculate total initial letter count
total_initial_letters = sum(initial_letter_freq.values())

# Calculate percentages and create a Pandas DataFrame for the results
results = {
    'Initial Letter': [],
    'Occurrences': [],
    'Percentage': []
}
for initial_letter, freq in initial_letter_freq.items():
    percentage = (freq / total_initial_letters) * 100
    results['Initial Letter'].append(initial_letter)
    results['Occurrences'].append(freq)
    results['Percentage'].append(percentage)

df_results = pd.DataFrame(results)

# Save the results as an Excel file
output_excel_path = 'Bengali_Initial_Letter_Percentage.xlsx'
df_results.to_excel(output_excel_path, index=False)
print(f"Percentage of occurrence of Bengali initial letters saved to {output_excel_path}")
```
